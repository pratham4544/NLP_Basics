```python
# NER model
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

# show NER results
from spacy import displacy
```

```python
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
pipe = pipeline("ner", model=model, tokenizer=tokenizer)
```

```python
text = "Hello! I'm Max. I'd like to book a flight to South Korea."

ner_results = pipe(text)
print(ner_results)
```

```python
def from_ner_results_to_displacy(text, ner_results):
  d_result = { "text": text, "title": None }
  ents = []
  current_entity = None
  for ent in ner_results:
    if "B-" in ent["entity"]:
      if current_entity != None:
        ents.append(current_entity)
      entity_label = ent["entity"][2:]
      current_entity = {
        "label": entity_label,
        "start": ent["start"],
        "end": ent["end"]
      }
    if "I-" in ent["entity"]:
      current_entity["end"] = ent["end"]
  if current_entity != None:
    ents.append(current_entity)
  d_result["ents"] = ents
  return d_result
```

```python
text = "Hello! I'm Max. I'd like to book a flight to South Korea."
ner_results = pipe(text)
d_displacy = from_ner_results_to_displacy(text, ner_results)
print(d_displacy)
```

The given `ClassLabel` defines Named Entity Recognition (NER) tags used in sequence labeling tasks to identify and classify entities in text. Here's a breakdown of the tags:

### **Tag Definitions**
1. **`O` (Outside)**:
   - Tokens that are not part of any named entity are tagged as `O`.
   - Example: *"I live in"* — "I" and "live" would be tagged as `O`.

2. **`B-PER` (Begin-Person)**:
   - The beginning of a person's name.
   - Example: *"Mahatma Gandhi"* — "Mahatma" would be tagged as `B-PER`.

3. **`I-PER` (Inside-Person)**:
   - The continuation of a person's name.
   - Example: *"Mahatma Gandhi"* — "Gandhi" would be tagged as `I-PER`.

4. **`B-ORG` (Begin-Organization)**:
   - The beginning of an organization name.
   - Example: *"OpenAI Research Lab"* — "OpenAI" would be tagged as `B-ORG`.

5. **`I-ORG` (Inside-Organization)**:
   - The continuation of an organization name.
   - Example: *"OpenAI Research Lab"* — "Research Lab" would be tagged as `I-ORG`.

6. **`B-LOC` (Begin-Location)**:
   - The beginning of a location name (e.g., city, country, or landmark).
   - Example: *"New York City"* — "New" would be tagged as `B-LOC`.

7. **`I-LOC` (Inside-Location)**:
   - The continuation of a location name.
   - Example: *"New York City"* — "York City" would be tagged as `I-LOC`.

### **Why Use `B-` and `I-` Tags?**
- **Boundary Clarity**: The `B-` (Begin) and `I-` (Inside) tags help in distinguishing adjacent entities of the same type.
  - Example: In *"John and Mary are in Paris"*, "John" and "Mary" are separate `PER` entities, so they are tagged `B-PER`.
- **Hierarchical Entities**: They allow distinguishing multi-token entities (e.g., "New York City") from single-token entities (e.g., "Paris").

### **Usage Context**
This tagging scheme is part of the **IOB format** (Inside-Outside-Begin), commonly used in NLP tasks to label tokens for training NER models. Libraries like **spaCy**, **Hugging Face Transformers**, and **NLTK** rely on similar schemes for entity recognition tasks.

```python
## Creating model for token classification

import torch.nn as nn

from transformers import XLMRobertaConfig
from transformers.modeling_outputs import TokenClassifierOutput
from transformers.models.roberta.modeling_roberta import RobertaModel, RobertaPreTrainedModel

```

```python
class XLMRobertaTokenlassification(RobertaPreTrainedModel):
  config_class = XLMRobertaConfig

  def __init__(self, config):
    super().__init__(config)

    self.num_labels = config.num_labels
    #load model body
    self.roberta = RobertaModel(config, add_pooling_layer=False)
    # set up token classification head
    self.dropout = nn.Dropout(config.hidden_dropout_prob)
    self.classifier = nn.Linear(config.hidden_size, config.num_labels)
    # load & initiate weights
    self.init_weights()

  def forward(self, input_ids=None, attention_mask=None, token_type_ids=None, labels=None, **kwargs):
    # 3 use modebl body to get encoder representation
    outputs = self.roberta(
        input_ids,
        attention_mask=attention_mask,
        token_type_ids=token_type_ids,
        **kwargs
    )
    seq_output = outputs[0]
    # apply classifier dropout & classifier head
    seq_output = self.dropout(seq_output)
    logits = self.classifier(seq_output)
    # calculate losses
    loss = None
    if labels is not None:
      loss_fct = nn.CrossEntropyLoss()
      loss = loss_fct(logits.view(-1, self.num_labels), labels.view(-1))
    # return model output object
    return TokenClassifierOutput(
        loss=loss,
        logits=logits,
        hidden_states=outputs.hidden_states,
        attentions=outputs.attentions
    )
```

```python
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizerFast, BertForTokenClassification, AdamW
from sklearn.model_selection import train_test_split
import ast

class LoanNERProcessor:
    def __init__(self, model_name='bert-base-uncased'):
        self.tokenizer = BertTokenizerFast.from_pretrained(model_name)
        self.label_map = {
            'O': 0,
            'B-CAR_LOAN': 1, 'I-CAR_LOAN': 2,
            'B-BIKE_LOAN': 3, 'I-BIKE_LOAN': 4,
            'B-PERSONAL_LOAN': 5, 'I-PERSONAL_LOAN': 6,
            'B-HOME_LOAN': 7, 'I-HOME_LOAN': 8,
            'B-EDUCATION_LOAN': 9, 'I-EDUCATION_LOAN': 10
        }
        self.inv_label_map = {v: k for k, v in self.label_map.items()}

    def tokenize_and_align_labels(self, text, entities, max_length=128):
        # Safely convert entities
        if isinstance(entities, str):
            entities = ast.literal_eval(entities)

        # Tokenize with fixed max length
        encoding = self.tokenizer(
            text,
            return_offsets_mapping=True,
            padding='max_length',
            max_length=max_length,
            truncation=True
        )

        # Initialize labels
        labels = [self.label_map['O']] * max_length

        # Align entities
        for entity in entities:
            start, end, label = entity['start'], entity['end'], entity['label']

            for idx, offset in enumerate(encoding['offset_mapping'][:max_length]):
                if offset[0] >= start and offset[1] <= end:
                    labels[idx] = (self.label_map[f'B-{label}']
                                   if offset[0] == start
                                   else self.label_map[f'I-{label}'])

        return encoding, labels

def prepare_dataset(df):
    processor = LoanNERProcessor()

    # Prepare data
    input_ids, attention_masks, label_list = [], [], []

    for _, row in df.iterrows():
        encoding, labels = processor.tokenize_and_align_labels(
            row['text'],
            row['entities']
        )

        input_ids.append(encoding['input_ids'])
        attention_masks.append(encoding['attention_mask'])
        label_list.append(labels)

    # Convert to tensors
    input_ids = torch.tensor(input_ids)
    attention_masks = torch.tensor(attention_masks)
    labels = torch.tensor(label_list)

    return input_ids, attention_masks, labels, processor

def main():
    # Load dataset
    df = pd.read_csv('loan_ner_dataset.csv')

    # Prepare dataset
    input_ids, attention_masks, labels, processor = prepare_dataset(df)

    # Split data
    (
        train_input_ids, val_input_ids,
        train_attention_masks, val_attention_masks,
        train_labels, val_labels
    ) = train_test_split(
        input_ids, attention_masks, labels,
        test_size=0.2,
        random_state=42
    )

    # Create DataLoaders
    train_dataset = torch.utils.data.TensorDataset(
        train_input_ids,
        train_attention_masks,
        train_labels
    )
    val_dataset = torch.utils.data.TensorDataset(
        val_input_ids,
        val_attention_masks,
        val_labels
    )

    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=16)

    # Model initialization
    model = BertForTokenClassification.from_pretrained(
        'bert-base-uncased',
        num_labels=len(processor.label_map)
    )

    # Device setup
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)

    # Training
    optimizer = AdamW(model.parameters(), lr=5e-5)
    model.train()

    for epoch in range(3):
        total_loss = 0
        for batch in train_loader:
            optimizer.zero_grad()

            input_ids, attention_mask, batch_labels = [b.to(device) for b in batch]

            outputs = model(
                input_ids,
                attention_mask=attention_mask,
                labels=batch_labels
            )

            loss = outputs.loss
            total_loss += loss.item()

            loss.backward()
            optimizer.step()

        print(f'Epoch {epoch+1}/3 Loss: {total_loss/len(train_loader)}')

    # Prediction example
    test_texts = [
        "I want to apply for a home loan",
        "Need an education loan for my studies",
        "Looking for a bike loan"
    ]

    model.eval()
    for text in test_texts:
        encoding = processor.tokenizer(
            text,
            return_tensors='pt',
            truncation=True,
            padding=True
        )

        with torch.no_grad():
            outputs = model(
                encoding['input_ids'].to(device),
                attention_mask=encoding['attention_mask'].to(device)
            )

        predictions = torch.argmax(outputs.logits, dim=2)[0]
        print(f"\nText: {text}")
        print("Predictions:", [processor.inv_label_map[pred.item()] for pred in predictions])

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizerFast, BertForTokenClassification, AdamW
from sklearn.model_selection import train_test_split
import ast

class LoanNERProcessor:
    def __init__(self, model_name='bert-base-uncased'):
        self.tokenizer = BertTokenizerFast.from_pretrained(model_name)
        self.label_map = {
            'O': 0,
            'B-CAR_LOAN': 1, 'I-CAR_LOAN': 2,
            'B-BIKE_LOAN': 3, 'I-BIKE_LOAN': 4,
            'B-PERSONAL_LOAN': 5, 'I-PERSONAL_LOAN': 6,
            'B-HOME_LOAN': 7, 'I-HOME_LOAN': 8,
            'B-EDUCATION_LOAN': 9, 'I-EDUCATION_LOAN': 10
        }
        self.inv_label_map = {v: k for k, v in self.label_map.items()}

    def tokenize_and_align_labels(self, text, entities, max_length=128):
        # Safely convert entities
        if isinstance(entities, str):
            entities = ast.literal_eval(entities)

        # Tokenize with fixed max length
        encoding = self.tokenizer(
            text,
            return_offsets_mapping=True,
            padding='max_length',
            max_length=max_length,
            truncation=True
        )

        # Initialize labels
        labels = [self.label_map['O']] * max_length

        # Align entities
        for entity in entities:
            start, end, label = entity['start'], entity['end'], entity['label']

            for idx, offset in enumerate(encoding['offset_mapping'][:max_length]):
                if offset[0] >= start and offset[1] <= end:
                    labels[idx] = (self.label_map[f'B-{label}']
                                   if offset[0] == start
                                   else self.label_map[f'I-{label}'])

        return encoding, labels

def prepare_dataset(df):
    processor = LoanNERProcessor()

    # Prepare data
    input_ids, attention_masks, label_list = [], [], []

    for _, row in df.iterrows():
        encoding, labels = processor.tokenize_and_align_labels(
            row['text'],
            row['entities']
        )

        input_ids.append(encoding['input_ids'])
        attention_masks.append(encoding['attention_mask'])
        label_list.append(labels)

    # Convert to tensors
    input_ids = torch.tensor(input_ids)
    attention_masks = torch.tensor(attention_masks)
    labels = torch.tensor(label_list)

    return input_ids, attention_masks, labels, processor

def main():
    # Load dataset
    df = pd.read_csv('loan_ner_dataset.csv')

    # Prepare dataset
    input_ids, attention_masks, labels, processor = prepare_dataset(df)

    # Split data
    (
        train_input_ids, val_input_ids,
        train_attention_masks, val_attention_masks,
        train_labels, val_labels
    ) = train_test_split(
        input_ids, attention_masks, labels,
        test_size=0.2,
        random_state=42
    )

    # Create DataLoaders
    train_dataset = torch.utils.data.TensorDataset(
        train_input_ids,
        train_attention_masks,
        train_labels
    )
    val_dataset = torch.utils.data.TensorDataset(
        val_input_ids,
        val_attention_masks,
        val_labels
    )

    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=16)

    # Model initialization
    model = BertForTokenClassification.from_pretrained(
        'bert-base-uncased',
        num_labels=len(processor.label_map)
    )

    # Device setup
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)

    # Training
    optimizer = AdamW(model.parameters(), lr=5e-5)
    model.train()

    for epoch in range(3):
        total_loss = 0
        for batch in train_loader:
            optimizer.zero_grad()

            input_ids, attention_mask, batch_labels = [b.to(device) for b in batch]

            outputs = model(
                input_ids,
                attention_mask=attention_mask,
                labels=batch_labels
            )

            loss = outputs.loss
            total_loss += loss.item()

            loss.backward()
            optimizer.step()

        print(f'Epoch {epoch+1}/3 Loss: {total_loss/len(train_loader)}')

    # Save the model and tokenizer
    model.save_pretrained('/content/loan_ner_model')
    processor.tokenizer.save_pretrained('/content/loan_ner_model')

    # Load the model and tokenizer from local directory for prediction
    loaded_model = BertForTokenClassification.from_pretrained('/content/loan_ner_model')
    loaded_tokenizer = BertTokenizerFast.from_pretrained('/content/loan_ner_model')

    loaded_model.to(device)
    loaded_model.eval()

    # Example user input prediction
    user_input = input("Enter your query: ")
    encoding = loaded_tokenizer(
        user_input,
        return_tensors='pt',
        truncation=True,
        padding=True
    )

    with torch.no_grad():
        outputs = loaded_model(
            encoding['input_ids'].to(device),
            attention_mask=encoding['attention_mask'].to(device)
        )

    predictions = torch.argmax(outputs.logits, dim=2)[0]
    print(f"Predictions for '{user_input}':", [processor.inv_label_map[pred.item()] for pred in predictions])

if __name__ == "__main__":
    main()
```
